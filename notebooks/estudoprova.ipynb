```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
raw = pd.read_excel('../trabalho/Space.xlsx', header=0, usecols=(2, 3, 4, 5, 6, 7, 8)).squeeze("columns")

raw['Datum'] = pd.to_datetime(raw['Datum'], dayfirst=True)
```

```python
raw['Location'].fillna('Location Unkown', inplace=True)

raw.head(30)
```

```python
search_for = ['Brazil']

raw[raw['Location'].str.contains('|'.join(search_for))]

```

```python
# Tratamento de erros

raw['Rocket'].fillna(0, inplace=True)

# filtros 

filtro_data = (raw['Datum'] > datetime(1950, 10, 1)) & (raw['Datum'] < datetime(2010, 10, 1))

filtro_missao = raw['Status Mission'] == 'Success'

filtro = filtro_data & filtro_missao

result = raw[filtro].groupby(['Company Name', 'Location']).agg(
    soma=('Rocket', 'sum'),
    media=('Rocket', 'mean')
)


result['soma'].idxmax()
```

```python
# add new column


res = pd.cut(raw['Rocket'], bins=[0, 25, 50, 75, raw['Rocket'].max()], include_lowest=True, labels=['baixo', 'medio', 'alto', 'muito alto'])
# res.rename(columns={"Rocket": "Class"}, inplace=True)
res2 = pd.concat([raw, res], axis=1)

res2.columns.values[7] = 'Putz'

res2
```
